```python
import requests
from PIL import Image
from io import BytesIO
```

```python
base_url = 'https://randomfox.ca/floof'
```

```python
def get_fox():

  try:
    response =  requests.get(base_url)
    response =  requests.get(base_url)
    if response.status_code == 200:
      fox_data = response.json()
      return fox_data
    else:
      print(response.status_code)
  except:
    print('Check your url and code again')


```

```python
fox_url_image = 'https://randomfox.ca/images/21.jpg'
def get_fox_image():
  try:
    response = requests.get(fox_url_image)
    img = Image.open(BytesIO(response.content))
    return img
  except Exception as e:
    print("cannot open image")

```
